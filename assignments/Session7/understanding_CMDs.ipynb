Below is the Gaia colour-magnitude diagram (CMD) for NGC 7789. The Gaia spacecraft measures the brightness, colour, and distance to stars with exquisite accuracy. Because Gaia also measures distances to stars and their proper-motion I have used distances and space motion to throw out stars that are in the line of sight to NGC 7789, but not members of the cluster. This makes the CMD very clean indeed.

To understand the features we can see in the CMD, we need to look at some evolutionary models, and remember what we have learnt in PHY213 about the evolution of stars. Below is an isochrone from the [MIST](https://ui.adsabs.harvard.edu/abs/2016ApJ...823..102C/abstract) stellar models. An isochrone is a plot of the locations of stars with a range of different masses, but all at the same age, so this is not quite the same as an *evolutionary track*, which would plot the location of a single star as it evolves over time. 

At the age of NGC 7789, only stars with masses greater than a few Solar masses have evolved off the main sequence, so we should expect to see some of the late evolutionary phases appropriate for higher mass stars in the isochrone. 

<img src="isochrone.png" style="margin: 0px" width=500px/>

1. **Main Sequence:** core Hydrogen burning. The hook to the left at MS turn-off is caused by the contraction of the star when Hydrogen burning stops. This is only seen in high mass stars (> 1$M_{\odot}$) because high mass stars have fully convective cores so can exhaust all the Hydrogen within the core.
2. **Sub-giant branch:** shell Hydrogen burning, radiative envelope.
3. **Giant branch:** shell Hydrogen burning, convective envelope.
4. **Core Helium Burning:** Helium core has ignited.

## Real CMDs

Comparing the isochrone with the actual CMD for NGC 7789, it's actually quite hard to match up the various phases to the stars in the CMD. Hopefully the Main-sequence is fairly clear, but which are the sub-giant stars? Or the giants?

To understand the observed CMD there is a bit of information we are missing, which is **how long** each phase of stellar evolution lasts! We might expect to find very few stars that are currently in short-lived phases of stellar evolution, whereas long-lived phases should be well populated with stars. To illustrate this, I have annotated the theoretical isochrone with a black line, where the thickness of the line crudely represents how many stars we might expect to see in that phase.

<img src="isochrone 2.png" style="margin: 0px" width=500px/>

We can see that the Main Sequence is well populated. There is no black line covering the sub-giant branch. That is because, for massive-stars, the Helium core cannot support it's own weight and collapses, causing a **very rapid transition** to the giant branch. As a result, we shouldn't expect to see any stars on the sub-giant branch. This feature of CMDs in open clusters is called the **Hertzsprung Gap**.

The giant branch itself is relatively short-lived, but we might expect to see a few giant branch stars. The planetary nebula phase is incredibly short, and we shouldn't see any stars in this region. White dwarfs are long lived, but they are too faint to be seen in our images.

### The Red Clump

What's the black blob near the horizontal branch, I hear you ask? Stars tend to ignite their Helium cores at very **similar core masses**. In turn, the luminosity of the whole star is controlled by the core mass, which leads to all these stars having very similar luminosities and temperatures. Therefore we expect a large number of stars with different masses to gather together in the same place in the CMD. This is known as the **red clump**, and consists of stars that are burning Helium in their cores.
